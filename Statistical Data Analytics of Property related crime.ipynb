```python
crime_india['Value_of_Property_Stolen']=stats.norm(scale=1, loc=0).rvs(2449)

#plotting a histogram
ax = sns.distplot(crime_india['Value_of_Property_Stolen'], bins=50,
                  kde=True,
                  color='yellow',
                  hist_kws={"linewidht": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')

plt.show()
```

```python
# Central tendency is a statistical concept that refers to a single value that represents the center or typical value of a distribution of data. It is used to describe the central or average value of a dataset, and it indicates where the data is clustering around.
# The three common measures of central tendency are the mean, median, and mode.
# Mean: The mean is the sum of all the values in a dataset divided by the number of values. It is also known as the arithmetic average. The mean is sensitive to extreme values, also known as outliers.
# Median: The median is the middle value of a dataset when it is sorted in order. It is the value below which 50% of the data lies and above which 50% of the data lies. The median is robust to extreme values.
# Mode: The mode is the value that occurs most frequently in a dataset. There may be more than one mode or no mode in a dataset.
# Central tendency provides a useful summary of the data in a single value, making it easier to interpret and compare different datasets.However, it is important to note that central tendency alone may not provide a complete picture of the dataset, and it should always be used in conjunction with other descriptive statistics and visualization techniques.

# Standard deviation is a statistical measure that quantifies the amount of variation or dispersion in a set of data. It measures how far the data is spread out from the mean (average) of the data set. A low standard deviation indicates that the data is clustered around the mean, while a high standard deviation indicates that the data is more spread out.
# Standard deviation is useful in data analysis because it provides a way to measure the spread of the data and identify outliers. For example, if the standard deviation is large, this may indicate that there is a lot of variation in the data, and the mean may not accurately represent the center of the distribution. In such cases, other measures, such as the median or interquartile range, may be more appropriate.
# Standard deviation can also be used to compare the variability of different data sets. For example, if two data sets have similar means but different standard deviations, this may indicate that one data set is more spread out than the other.



# The harmonic mean is a statistical measure that is used to calculate the average of a set of numbers. It is calculated by taking the reciprocal of each number in the set, averaging these reciprocals, and then taking the reciprocal of the result.
# In other words, if we have a set of numbers x1, x2, ..., xn, the harmonic mean can be calculated as:
# Harmonic Mean = n / (1/x1 + 1/x2 + ... + 1/xn)
# where n is the number of observations.
# The harmonic mean is useful in situations where the rate of change is important, or where the observations have a reciprocal relationship. For example, it can be used to calculate the average speed of a car over a journey, taking into account any changes in speed that occurred during the journey. The harmonic mean is also used in financial analysis, such as calculating the average return on investments that have been held for different periods of time.
# Compared to other measures of central tendency such as the arithmetic mean or median, the harmonic mean gives more weight to smaller values in the set, making it useful in situations where outliers or extreme values are present. However, it is less commonly used than other measures of central tendency because it can be more difficult to interpret and understand.

# Pearson typically refers to the Pearson correlation coefficient, which is a statistical measure that quantifies the strength and direction of the linear relationship between two variables.
# The Pearson correlation coefficient, often denoted as "r", ranges from -1 to +1, where a value of -1 indicates a perfect negative correlation (i.e., as one variable increases, the other variable decreases), a value of +1 indicates a perfect positive correlation (i.e., as one variable increases, the other variable also increases), and a value of 0 indicates no correlation (i.e., there is no linear relationship between the variables).
# The Pearson correlation coefficient is often used in various fields, including psychology, sociology, economics, and engineering, to name a few. It is particularly useful when researchers want to examine the relationship between two quantitative variables, such as height and weight, or education level and income. By calculating the Pearson correlation coefficient, researchers can determine the strength and direction of the relationship between these variables and use this information to draw meaningful conclusions about their findings.

# Quartiles are a statistical measure that divides a data set into four equal parts. The quartiles are calculated by first ordering the data from smallest to largest and then dividing it into four equally sized groups. The values that divide the data into these groups are called quartiles.
# The three quartiles that divide the data into four equal parts are called the first quartile (Q1), the second quartile (Q2), and the third quartile (Q3). The second quartile is also known as the median.
# Quartiles are useful because they can provide information about the distribution of a data set. For example, if the first quartile is much smaller than the third quartile, this indicates that the data is skewed to the right (i.e., there are more high values in the data set than low values). If the first and third quartiles are roughly the same size, this indicates that the data is roughly symmetrical.
# Quartiles are also useful in data analysis because they can be used to calculate other statistical measures, such as the interquartile range (IQR). The IQR is the range of values that lie between the first and third quartiles, and it provides information about the spread of the middle 50% of the data set. The IQR can be used to identify outliers (values that are significantly different from the rest of the data) and to compare the variability of different data sets.

# Skewness is a statistical measure that quantifies the degree of asymmetry in a probability distribution. In other words, it measures how much a distribution deviates from symmetry. A distribution is considered to be symmetric if it is evenly distributed around its mean and has no skewness.
# Skewness can be positive, negative, or zero. Positive skewness indicates that the distribution has a longer right tail (i.e., there are more high values in the data set than low values), while negative skewness indicates that the distribution has a longer left tail (i.e., there are more low values in the data set than high values). A skewness of zero indicates that the distribution is symmetric.
# Skewness is useful in data analysis because it can provide insights into the underlying data. For example, if a distribution is positively skewed, this might indicate that there are a few very high values that are causing the skewness. On the other hand, if a distribution is negatively skewed, this might indicate that there are a few very low values that are causing the skewness.
# Skewness can also affect the interpretation of other statistical measures, such as the mean and standard deviation. When a distribution is skewed, the mean is typically pulled in the direction of the skewness. This means that the mean may not accurately represent the center of the distribution, and other measures such as the median may be more appropriate. Additionally, when a distribution is skewed, the standard deviation may not accurately represent the spread of the distribution, and other measures such as the interquartile range may be more appropriate

# Kurtosis is a statistical measure that quantifies the degree of peakedness or flatness of a probability distribution compared to the normal distribution. The normal distribution has a kurtosis of 3, so distributions with kurtosis greater than 3 are more peaked (i.e., have a sharper peak) than the normal distribution, while distributions with kurtosis less than 3 are flatter than the normal distribution.
# Kurtosis can be positive or negative. Positive kurtosis indicates a distribution that is more peaked than the normal distribution, while negative kurtosis indicates a distribution that is flatter than the normal distribution. A kurtosis of 3 indicates that the distribution is normal (i.e., it has the same shape as the normal distribution).
# Kurtosis is useful in data analysis because it can provide insights into the underlying data. For example, a distribution with high positive kurtosis may indicate that there are a few extreme values that are causing the peak, while a distribution with high negative kurtosis may indicate that the data is spread out more widely than in a normal distribution.
# Kurtosis can also affect the interpretation of other statistical measures, such as the mean and standard deviation. When a distribution has high positive kurtosis, the mean may be pulled towards the peak, which means that it may not accurately represent the center of the distribution. Additionally, when a distribution has high positive kurtosis, the standard deviation may be larger than in a normal distribution, which means that it may not accurately represent the spread of the distribution. In such cases, other measures, such as the median or interquartile range, may be more appropriate.

# A histogram is a graphical representation of the distribution of a set of numerical data. It consists of a set of adjacent rectangular bars, where each bar represents an interval or "bin" of the data, and the height of the bar represents the count or frequency of data points falling into that bin.
# Here are the steps to read a histogram:
# Look at the x-axis: The x-axis of a histogram represents the range of the data values. The axis is divided into intervals or bins, and the width of each bin represents the size of the interval. The values at the edges of the bins are called "bin boundaries" or "bin edges".
# Look at the y-axis: The y-axis of a histogram represents the count or frequency of data points falling into each bin. The height of each bar corresponds to the frequency or count of data points in the corresponding bin.
# Interpret the shape: The shape of the histogram provides information about the distribution of the data. A symmetric, bell-shaped histogram indicates a normal distribution, while a histogram that is skewed to the left or right indicates a skewed distribution. A histogram with multiple peaks indicates a bimodal or multimodal distribution.
# Look for outliers: Outliers are data points that are much larger or smaller than the other data points in the dataset. They can be identified as bars that are much taller or shorter than the other bars in the histogram.
# Interpret the density: The density of the histogram gives an indication of the relative proportion of data points in each bin. A high-density histogram indicates that there are many data points in that range, while a low-density histogram indicates that there are fewer data points in that range.
# By following these steps, you can interpret the information presented in a histogram and gain insights into the distribution of a dataset.

# Binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials, where each trial has a probability of success, denoted by "p". The distribution is characterized by two parameters: the number of trials, denoted by "n", and the probability of success, denoted by "p".
# The graph of a binomial distribution can be represented using a histogram or a probability mass function (PMF) plot. The horizontal axis of the graph represents the number of successes, while the vertical axis represents the probability of that number of successes occurring.
# To read a graph of a binomial distribution, follow these steps:
# Identify the parameters: The graph should indicate the values of n and p used to generate the distribution.
# Observe the shape: The shape of the distribution depends on the values of n and p. As n increases or p gets closer to 0.5, the distribution becomes more symmetric and bell-shaped, resembling a normal distribution. When n is small or p is very small or very large, the distribution becomes more skewed.
# Look at the peak: The peak of the distribution represents the most likely number of successes.
# Observe the spread: The spread of the distribution is determined by the values of n and p. As n increases or p gets closer to 0.5, the spread of the distribution decreases.
# Look for outliers: Outliers in a binomial distribution occur when the number of successes is much higher or lower than expected. These outliers may indicate unusual or unexpected results.
# Overall, the graph of a binomial distribution provides a visual representation of the probabilities of different numbers of successes

# Poisson data is a type of discrete data that arises from counting the number of occurrences of a certain event within a fixed time interval or within a fixed area or volume. For example, the number of customers arriving at a store in a given hour, the number of cars passing a certain point on a road in a day, or the number of defects in a batch of products are all examples of Poisson data.
# The Poisson distribution is a probability distribution that models the number of occurrences of rare events in a fixed interval or region. The Poisson distribution is characterized by a single parameter, denoted by "lambda" (), which represents the expected number of occurrences in the given interval or region.
# The graph of a Poisson distribution can be represented using a histogram or a probability mass function (PMF) plot. The horizontal axis of the graph represents the number of occurrences, while the vertical axis represents the probability of that number of occurrences occurring.
# To interpret a graph of Poisson data, follow these steps:
# Identify the parameter: The graph should indicate the value of lambda () used to generate the Poisson distribution.
# Observe the shape: The shape of the distribution depends on the value of lambda. As lambda increases, the distribution becomes more symmetric and bell-shaped, resembling a normal distribution.
# Look at the peak: The peak of the distribution represents the most likely number of occurrences.
# Observe the spread: The spread of the distribution is determined by the value of lambda. As lambda increases, the spread of the distribution increases as well.
# Look for outliers: Outliers in a Poisson distribution occur when the number of occurrences is much higher or lower than expected. These outliers may indicate unusual or unexpected results.
# Overall, the graph of a Poisson distribution provides a visual representation of the probabilities of different numbers of occurrences, and can be a useful tool for understanding the behavior of a system with rare events


# Lognormal distribution is a continuous probability distribution of a random variable whose logarithm is normally distributed. In other words, a lognormal distribution is a probability distribution of a random variable whose logarithm has a normal distribution.
# The lognormal distribution is used to model data that is skewed to the right, where most of the values are clustered on the left-hand side of the distribution and there are a few extreme values on the right-hand side. It is commonly used in fields such as economics, finance, and biology to model variables such as income, asset prices, and sizes of biological organisms.

# The lognormal distribution is characterized by two parameters: mu and sigma. Mu is the mean of the logarithm of the data, and sigma is the standard deviation of the logarithm of the data.
# In a lognormal distribution graph, the x-axis represents the values of the random variable, and the y-axis represents the probability density of those values. The shape of the distribution will depend on the values of mu and sigma.

# A pair plot, also known as a scatter plot matrix, is a plot that allows us to visualize pairwise relationships between variables in a dataset. A pair plot consists of multiple scatterplots arranged in a grid, with each plot showing the relationship between a pair of variables.
# The pair plot shows the scatterplot between each pair of variables in the dataset, and a histogram of each variable along the diagonal of the grid. By default, the pair plot plots the numerical variables of the dataset, but it can also show the categorical variables with different colors or markers.
# To read a pair plot, we need to look at the scatterplots in each cell of the grid. Each scatterplot shows the relationship between the two variables on its x and y axes. We need to look for patterns, such as linear or non-linear relationships, clusters, outliers, and correlations. We can also look at the histograms on the diagonal to see the distribution of each variable.
# In addition, we can use the hue parameter of the sns.pairplot() function to add a categorical variable to the plot. This allows us to see how the relationship between the numerical variables changes across different categories. We can also use the kind parameter to change the type of plot used in the diagonal, such as a kernel density estimate (KDE) plot or a rug plot.
# Overall, pair plots are a useful tool for exploratory data analysis (EDA) and can
#  help us gain insights into the relationships between variables in a dataset.


# A heatmap is a graphical representation of data that uses color-coding to represent values in a matrix. Heatmaps are commonly used in data visualization to show patterns and relationships between variables in a dataset.
# A heatmap can be created in Python using the seaborn library. To create a heatmap, we need a matrix of values and row and column labels for the matrix. We can then use the sns.heatmap() function to plot the matrix.
# In a heatmap, each value in the matrix is assigned a color based on its numerical value, with high values represented by a darker color and low values represented by a lighter color. The color scale used in the heatmap can be customized using the cmap parameter of the sns.heatmap() function.
# To read a heatmap, we need to look at the color coding of each cell in the matrix. We can look for patterns in the color coding, such as clusters or gradients, that may indicate relationships or trends between the variables. We can also look for outliers or values that stand out from the rest of the data.
# Heatmaps can be used to visualize a variety of data types, such as correlation matrices, gene expression data, or geographic data. They are particularly useful when working with large datasets or when trying to identify patterns or relationships between many variables.

# Correlation is a statistical measure that indicates how strongly two variables are related to each other. It is a value that ranges from -1 to +1, with a value of -1 indicating a perfect negative correlation (meaning that as one variable increases, the other decreases), a value of +1 indicating a perfect positive correlation (meaning that as one variable increases, the other also increases), and a value of 0 indicating no correlation (meaning that there is no relationship between the two variables).
# Correlation is commonly used to investigate the relationship between different variables in a dataset, such as the relationship between a person's height and weight, or between the price of a product and the number of units sold. By understanding the strength and direction of the correlation between variables, researchers can gain insights into the nature of the relationship and make predictions about how changes in one variable may affect the other.









```
